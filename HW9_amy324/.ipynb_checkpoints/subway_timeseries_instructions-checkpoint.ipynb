### Now we have a general idea about our data, let's start our analysis on this timeseries to find out if we notice some different behaviour.

````python
ridesPerWeek = np.nansum(allStationRides, axis=0)
````

````python
# plotting a sum of all stations and all ridership types during the period
pl.figure(figsize=(10,5))

pl.plot(dateRange, np.log10(ridesPerWeek))
    
pl.xticks(rotation=80, fontsize=12)
pl.yticks(fontsize=10)

pl.xlabel("Weeks from 21/05/2010 to 21/02/2014", fontsize=12)
pl.ylabel("Number of Rides - log10", fontsize=12)
pl.title("Number of rides per week (log10) - 21/05/2010 to 21/02/2014", fontsize=12)
````

#### Figure 2a: We can notice here that between September, 2012 and January, 2013 something happened and made the number of rides drop in one day. Let's investigate it more.

### We will draw a mean line and a threshold to try to identify events that make the rides go more than 3 standart deviations away from the mean.

````python
np.nanmean(ridesPerWeek)
````

````python
# Creating threshold and mean line
print('Mean rides per week:', np.nanmean(ridesPerWeek))
print('Standart deviation of rides per week:', np.nanstd(ridesPerWeek))
thresholds = [np.nanmean(ridesPerWeek) + 3 * np.nanstd(ridesPerWeek), 
              np.nanmean(ridesPerWeek) - 3 * np.nanstd(ridesPerWeek)]
print ('Thresholds:', thresholds)
````

````python
pl.figure(figsize=(10,6))

pl.plot(dateRange, np.log10(ridesPerWeek), label = 'Rides')
pl.plot([dateRange[0], dateRange[-1]],[np.log10(thresholds[0]), np.log10(thresholds[0])], '-r', linewidth=2.0)
pl.plot([dateRange[0], dateRange[-1]],[np.log10(thresholds[1]), np.log10(thresholds[1])], 
        '-r', linewidth=2.0, label = 'Threshold')
outlier = np.where(ridesPerWeek<thresholds[1])
pl.plot(dateRange[outlier], np.log10(ridesPerWeek[outlier]), 'rd', ms=8, label='Event detected')

pl.xticks(rotation=80, fontsize=10)
pl.xlabel("Weeks from 21/05/2010 to 21/02/2014", fontsize=12)
pl.ylabel("Number of Rides - log10", fontsize=12)
pl.title("Number of rides per week (log10) by ridership types - 21/05/2010 to 21/02/2014", fontsize=12)

pl.legend(loc='lower right', shadow=False, fontsize=10)
````

#### Figure 2b: We can notice that on the week of 10/19/2010 an event happened and decreased the number of rides. Researching about this range date, we can find that an important event that happened in NYC during this period is Hurricane Sandy (https://en.wikipedia.org/wiki/Hurricane_Sandy)

## Task 2: 
### Some of the time series are stationary, some have a complex structure, some show a downward trend: Identify the **ridership types**  that have steadily increased in popularity,  and that have steadily decreased. by how much? (e.g what is the ratio of usage in the first 10 and last 10 weeks)

### To find out wich ridership types have follow a trend or remain stationary, let's plot one by one and make a quick visual analysis about the plots.

````python
byStations = allRidesStations
````

````python
# Plotting the rolling mean, one by one, for every ridership type
pl.figure(figsize=(15,30))
pl.suptitle('Rolling Mean of rides per Ridership type (21/05/2010 - 21/02/2014)', fontsize=15)
pl.subplot(8,3,1)

for i in range(0, len(byStations)):
    #pl.tight_layout(pad=1.11, h_pad=1.8)
    pl.subplots_adjust(top = 0.96, wspace=0.4, hspace=0.75)
    pl.subplot(8,3,i+1)
    pl.plot(dateRange, pd.rolling_mean(byStations[i], 10))
    pl.title('Ridership type: %s ' %(ridetype[i]), fontsize=14)
    pl.xticks(rotation=70, fontsize=9)
    pl.xlabel('Weeks since 21/05/2010', fontsize=11)
    pl.ylabel('rolling mean - log10(rides)', fontsize=11)

````

#### Figure 4: we can see by this plot that there is some types of ridership that are missing information, others that follow a trend (like 14d), and others that remain stationary (like ada). To make us sure wich ones are Stationary or Non-Stationary, let's perform a ADFuller test, this test indicates us wich timeframe are more likely to be stationary or not.

````python

stationary = []
increase = {}
decrease = {}
nonStationary = []
err = []

# Extracting p-value and comparing with 5% significance level

for i in range(0, len(byStations)):
    try:
        adfTest = sm.tsa.adfuller(np.log10(byStations[i]))[1]
        if adfTest < 0.05:
            print(i, sm.tsa.adfuller(np.log10(byStations[i]))[1], ' Stationary')
            stationary.append(i)
        else:
            print(i, sm.tsa.adfuller(np.log10(byStations[i]))[1], ' Non-Stationary')
            ratio = byStations[i][-10:].sum()/byStations[i][:10].sum()
            nonStationary.append(i)
            if ratio > 1:
                increase[i] = ratio
            if ratio < 1:
                decrease[i] = ratio
    except LinAlgError:
        err.append(i)
        #print(i, ' Unavailable data')
        pass
````

### Now we performed the test, let's identify wich one is increasing and wich one is decreasing. We also will see on wich factor it is increasing or decreasing

````python
print ("The ridership types that has increased it's popularity are:")
for i in sorted(increase.keys()):
    print ("- Ridership type: %s (index=%d)  increased it's rides by a factor of: %f" %(ridetype[i], i, increase[i]))
````

````python
print ("The ridership types that has decreased it's popularity are:")
for i in sorted(decrease.keys()):
    print ("- Ridership type: %s (index=%d) decreased it's rides by a factor of: %f" %(ridetype[i], i, decrease[i]))
````

#### The type of ridership that increased more is Senior Citizen / Disabled
#### The type of ridership that decreased more is Monthly

````python
# Plotting the rolling mean, one by one, for every ridership type
pl.figure(figsize=(15,5))
pl.suptitle('Rolling Mean of top decreasing/increasing ratio Ridership Type(21/05/2010 - 21/02/2014)', fontsize=15)
pl.subplot(1,2,2)

count = 0
for i in [13, 16]:
    #pl.tight_layout(pad=1.11, h_pad=1.8)
    pl.subplots_adjust(top = 0.85, wspace=0.4, hspace=0.75)
    pl.subplot(1,2,1+count)
    pl.plot(dateRange, pd.rolling_mean(byStations[i], 10))
    pl.title('Ridership type: %s ' %(ridetype[i]), fontsize=14)
    pl.xticks(rotation=70, fontsize=9)
    pl.xlabel('Weeks since 21/05/2010', fontsize=11)
    pl.ylabel('rolling mean - log10(rides)', fontsize=11)
    count = count + 1
````

## Task 3:
### Several stations show a prominent annual periodicity. Identify the 4 **stations** (indentify them by the index of their location in the data cube) that show the most prominent periodic trend on an annual period (52 weeks). (Can you figure out what the periodic peak in rides is due to?)

* to find periodicity you can use fourier transforms. Look at the notebook on fourier transforms in https://github.com/fedhere/UInotebooks/tree/master/timeseries. If you choose to use the numpy package you may want to use the real fft function:  np.fft.rfft  
```
np.fft.rfft(mydata) 
```  

    will return the *real* part of the fourier transform (which is a *complex* object, and therefore it has a real and an *imaginary* component - that is just math. All you need to know is that you need the real part). But what you want is its absolute value: np.abs(). Then to get the x axis of your plot in frequency (1/week since the data is in weeks) you can use the function 

```
np.fft.rfftfreq(N, d) 
```     
     which takes as arguments N = the length of your time series (the number of datapoints) and d: the *sample spacing*, or 1.0/sampling_rate (in this case, with 1/week sampling, if you want the freq in units of week that is just 1!)

### Following the nstructions of Dr. Bianco, let's calculate fourier transforms to find out wich stations we can find more "power"

````python
byStations = allStationRides
byStations.shape
````

````python
f = np.abs(np.fft.rfft(byStations))
````

````python
f.shape
````

````python
N = 194
d = 1
freq = np.fft.rfftfreq(N, d)
period = 1/freq
````

````python
period
````

### We noticed here that the closest period from 52 weeks is 48.5 weeks. So let's find the max on this point on the timeframe for each station, and pick the highest 4

````python
fig = pl.figure(figsize=(16,8))
ax=fig.add_subplot(111)
for i in range(0, len(byStations)):
    ax.plot(freq[1:], f[i][1:], '-')
    ax.plot(freq[1:], f[i][1:], 'o', ms=3)
    
#ax.set_xticklabels([ "%s"%(1/f) for f in freq[1:]], rotation=70, fontsize = 15)
ax.set_xlabel('Period (weeks)', fontsize = 18)
ax.set_ylabel('Power', fontsize = 18)
````

#### Figure 6: This figure represent a plot of all stations and it's respective fourier 'power' on each time of it's timeframe

### The idea here is to find the highest powers on the 4th period (48.5 weeks). To do that, let's create a threshold and the stations that are over this threshold will be printed:

````python
power = []
# Here I was helped by Fernando Melchor (lmf445)
for i in range(len(byStations)):
    if ((f[i][4]>1500000) & (i!=0)):
        print ('Station: %d - Power: %f' %(i, f[i][4]))
        power.append(i)
print (power)
````

````python
fig = pl.figure(figsize=(16,8))
ax=fig.add_subplot(111)
for i in power:
    ax.plot(freq[1:], f[i][1:], '-')
    ax.plot(freq[1:], f[i][1:], 'o', ms=3)
    
ax.set_xticklabels([ "%s"%(1/f) for f in freq[1:]], rotation=70, fontsize = 15)
ax.set_xlabel('Period (weeks)', fontsize = 18)
ax.set_ylabel('Power', fontsize = 18)
````

#### Figure 7: This figure represent a plot of the most 4 'powerful' stations considering fourier 'power' on each time of it's timeframe

````python
# Plotting the rolling mean, one by one, for every ridership type
pl.figure(figsize=(10,5))
#pl.suptitle('Rolling Mean of top decreasing/increasing ratio Ridership Type(21/05/2010 - 21/02/2014)', fontsize=15)
pl.subplot(1,1,1)

count = 0
for i in power:
    #pl.tight_layout(pad=1.11, h_pad=1.8)
    pl.subplots_adjust(top = 0.8, wspace=0.4, hspace=0.75)
    pl.subplot(1,1,1)
    pl.plot(dateRange, pd.rolling_mean(log10(byStations[i]), 10), label = 'Station %d' %(i))
    pl.title('The 4 stations with more likely annual patterns', fontsize=16)
    pl.xticks(rotation=70, fontsize=9)
    pl.xlabel('Weeks since 21/05/2010', fontsize=11)
    pl.ylabel('rolling mean - log10(rides)', fontsize=11)
    pl.legend(loc='lower right', shadow=False, fontsize=9)
````
