`````python

stationary = []
increase = {}
decrease = {}
nonStationary = []
err = []

for i in range(0, len(byStations)):
    try:
        adfTest = sm.tsa.adfuller(np.log10(byStations[i]))[1]
        if adfTest < 0.05:
            print(i, sm.tsa.adfuller(np.log10(byStations[i]))[1], ' Stationary')
            stationary.append(i)
        else:
            print(i, sm.tsa.adfuller(np.log10(byStations[i]))[1], ' Non-Stationary')
            ratio = byStations[i][-10:].sum()/byStations[i][:10].sum()
            nonStationary.append(i)
            if ratio > 1:
                increase[i] = ratio
            if ratio < 1:
                decrease[i] = ratio
    except LinAlgError:
        err.append(i)
        #print(i, ' Unavailable data')
        pass
`````
