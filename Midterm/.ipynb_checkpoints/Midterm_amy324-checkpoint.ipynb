### 1. Download two months of citibike data 201506 and 201601

```python
# Function obtained from Dr. Bianco example
def getCitiBikeCSV(datestring):
    print ("Downloading", datestring)
    ### First I will check that it is not already there
    if not os.path.isfile(os.getenv("PUI2016") + "/" + datestring + "-citibike-tripdata.csv"):
        if os.path.isfile(datestring + "-citibike-tripdata.csv"):
            # if in the current dir just move it
            if os.system("mv " + datestring + "-citibike-tripdata.csv " + os.getenv("PUIDATA")):
                print ("Error moving file!, Please check!")
        #otherwise start looking for the zip file
        else:
            if not os.path.isfile(os.getenv("PUIDATA") + "/" + datestring + "-citibike-tripdata.zip"):
                if not os.path.isfile(datestring + "-citibike-tripdata.zip"):
                    os.system("curl -O https://s3.amazonaws.com/tripdata/" + datestring + "-citibike-tripdata.zip")
                ###  To move it I use the os.system() functions to run bash commands with arguments
                os.system("mv " + datestring + "-citibike-tripdata.zip " + os.getenv("PUIDATA"))
            ### unzip the csv 
            os.system("unzip " + os.getenv("PUIDATA") + "/" + datestring + "-citibike-tripdata.zip")
            ## NOTE: old csv citibike data had a different name structure. 
            if '2014' in datestring:
                os.system("mv " + datestring[:4] + '-' +  datestring[4:] + 
                          "\ -\ Citi\ Bike\ trip\ data.csv " + datestring + "-citibike-tripdata.csv")
            os.system("mv " + datestring + "-citibike-tripdata.csv " + os.getenv("PUIDATA"))
    ### One final check:
    if not os.path.isfile(os.getenv("PUIDATA") + "/" + datestring + "-citibike-tripdata.csv"):
        print ("WARNING!!! something is wrong: the file is not there!")

    else:
        print ("file in place, you can continue")
```

### 3. Find the zipcodes of citibike stations by reverse geocoding the coordinates 

```python
stat15 = df_15.copy()
stat16 = df_16.copy()
```

```python
stat15 = stat15.drop_duplicates('start station id')
stat16 = stat16.drop_duplicates('start station id')
```

```python
stat = pd.concat([stat15, stat16])
```

```python
print(len(stat15))
print(len(stat16))
print(len(stat))
```

```python
stat = stat.drop_duplicates('start station id')
print(len(stat))
```

```python
stat.columns
```

```python
# here is the dataframe with all station ID we need to find out by Google API
station = stat[[u'start station id',
       u'start station latitude',
       u'start station longitude']]
```

```python
station.head(10)
```

```python
key = 'Put your key here'
```

```python
zipcodes = []
```

```python
for i in range(0, len(station)):
    url = 'https://maps.googleapis.com/maps/api/geocode/json?latlng=' + str(station['start station latitude'].values[i]) + ',' \
          + str(station['start station longitude'].values[i]) + '&key=' + key
    #print(url)
    response = ulr.urlopen(url)
    data = response.read().decode("utf-8")
    data = json.loads(data)
    if len(data['results'][0]['address_components'][-1]['long_name']) == 5:
        zipcodes.append(data['results'][0]['address_components'][-1]['long_name'])
    else:
        zipcodes.append(data['results'][0]['address_components'][-2]['long_name'])
```

```python
len(station)
```

```python
len(zipcodes)
```

```python
zipcodes
```

```python
station['Zip Codes'] = zipcodes
```

```python
station.head()
```

### 5. Fit a line to ridership (number of rides over one of the 2 months of citibike data downloaded, start with 2015) vs income (total Adjusted gross income for the zip).

```python
countRide16.head()
```

```python
income_totals
```

```python
income_totals = income_totals[:-16]
```

```python
income_totals
```

```python
stations15 = countRide15.groupby([countRide15['Zip Codes']], as_index=False).count()
```

```python
stations15.drop(['usertype', 'gender'], axis=1, inplace=True)
```

```python
stations15['Zip Codes'] = stations15['Zip Codes'].astype(int)
```

```python
stations15 
```

```python
merged15 = pd.merge(stations15, income_totals, on = 'Zip Codes', how='inner')
merged15.columns = ['Zip Codes', 'Number of Rides', 'Adjusted Gross Income']
merged15.head()
```

```python
merged15
```

```python
fit15 = sm.OLS(merged15['Number of Rides'], merged15['Adjusted Gross Income'], missing='drop').fit()
```

```python
plt.figure(figsize=(8,8))

pl.plot(merged15['Adjusted Gross Income'], merged15['Number of Rides'],
         '.', ms = 8, alpha = 0.6, color = "SteelBlue"
        )
plt.plot(merged15['Adjusted Gross Income'], 
         fit15.predict(merged15['Adjusted Gross Income']), 
         'r-')

pl.title('Income x Num. of Rides - 201506')
pl.xlabel('Adjusted Gross Income', fontsize = 15)
pl.ylabel('Number of Rides', fontsize = 15)
```

```python
stations16 = countRide16.groupby([countRide16['Zip Codes']], as_index=False).count()
stations16.drop(['usertype', 'gender'], axis=1, inplace=True)
```

```python
stations16['Zip Codes'] = stations16['Zip Codes'].astype(int)
```

```python
plt.figure(figsize=(8,8))

pl.plot(merged16['Adjusted Gross Income'], merged16['Number of Rides'],
         '.', alpha = 0.6, color = "SteelBlue"
        )

plt.plot(merged16['Adjusted Gross Income'], 
         fit16.predict(merged16['Adjusted Gross Income']), 
         'r-')

pl.title('Income x Num. of Rides - 201601')
pl.xlabel('Adjusted Gross Income', fontsize = 15)
pl.ylabel('Number of Rides', fontsize = 15)
```

### 7. Fit a 2nd degree polynomial to the same data

```python
dataDF = pd.DataFrame()
dataDF['x'] = merged15['Adjusted Gross Income']
dataDF['x2'] = merged15['Adjusted Gross Income'] ** 2
dataDF['y'] = merged15['Zip Codes']
dataDF.head()
```

```python
fres = smf.ols(formula = 'y ~ x2 + x', data=dataDF).fit()
fres.summary()
```

```python
plt.figure(figsize=(8,8))

pl.plot(merged15['Adjusted Gross Income'], merged15['Number of Rides'],
         '.', ms = 8, alpha = 0.6, color = "SteelBlue"
        )
plt.plot(merged15['Adjusted Gross Income'], 
         fit15.predict(merged15['Adjusted Gross Income']), 
         'r-')

dfnew = pd.DataFrame()
dfnew['x'] = np.arange(merged15['Adjusted Gross Income'].min(), 
                       merged15['Adjusted Gross Income'].max(), 
                       0.1)

dfnew['x2'] = dfnew['x'] * dfnew['x']

pl.plot(dfnew['x'], 
        fres.predict(dfnew), 'g-')

pl.title('Income x Num. of Rides - 201506')
pl.xlabel('Adjusted Gross Income', fontsize = 15)
pl.ylabel('Number of Rides', fontsize = 15)

```
