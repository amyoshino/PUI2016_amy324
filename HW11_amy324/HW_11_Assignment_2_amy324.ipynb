### Now we downloaded the data, let's prepare it to merge in an one unique dataset

```python
# File names
files = []
name = []
for y in range(94, 100):
    files.append('zbp' + str(y) + 'totals.zip')
    name.append('19' + str(y))

for y in range(0, 10):
    files.append('zbp0' + str(y) + 'totals.zip')
    name.append('200' + str(y))

for y in range(10, 15):
    files.append('zbp' + str(y) + 'totals.zip')
    name.append('20' + str(y))
```

### Importing zipcode files and  merging with the data obtained above

```python
# Importing GeoJSON file

zipcodes = gpd.GeoDataFrame.from_file('nyc-zip-code-tabulation-areas-polygons.geojson')
```

```python
zipcodes = zipcodes[['postalCode', 'geometry']]
zipcodes.columns = [['zip', 'geometry']]
```

```python
zipcodes.head()
```

```python
# Our objective here is to merge all datasets in an unique one, indexed by zipcodes only from NY. The merge function
# can make it happen, using hte inner method.
# We saw before that some files has a column called est or EST. Let's make if statements to 
# handle those incompatibilities
i = 0
for fname in files:
    zf = zipfile.ZipFile(fname)
    df = pd.read_csv(zf.open(fname.replace('.zip','.txt')))
    
    # Read file and create a dataframe wit zipcode and establishment columns
    try:
        df = df[['ZIP', 'EST']]
        df = df.rename(columns={'ZIP': 'zip', 'EST': name[i]})
    except:
        df = df[['zip', 'est']]
        df = df.rename(columns={'zip': 'zip', 'est': name[i]})
        
    # Transforming in int for merging compatibility purpose
    zipcodes['zip'] = zipcodes['zip'].astype(int)
        
    zipcodes = pd.merge(zipcodes, df, how='inner', on='zip')

    i = i+1
```

```python
zipcodes.shape
```

```python
zipcodes = zipcodes.sort_values('zip', ascending=True)
zipcodes.index = zipcodes['zip']
```
